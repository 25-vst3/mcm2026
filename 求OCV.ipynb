$$OCV(SOC, T, SOH)
=
OCV_{ref}(SOC)
+
k_T(SOC)\bigl(T - T_{ref}\bigr)
+
\bigl(1 - SOH\bigr)\Delta OCV_{aging}(SOC),$$

以23摄氏度下的OCV(SOC)为基准曲线，考虑温度 $T$ 和电池寿命 $SOH$ 的影响

```python
import pandas as pd
import numpy as np

t_ref = 23 # 参考温度，摄氏度
discharging_23c = pd.read_csv('./data/OCV_SOC_23C_discharging.csv') # TODO 换成充放电的平均会不会更好？
discharging_45c = pd.read_csv('./data/OCV_SOC_45C_discharging.csv')

# 基准OCV曲线：23摄氏度下放电曲线，最终选用这个
def ocv_ref_23(soc, table):
    table.sort_values(by='SOC', inplace=True)
    res = np.interp(soc, table['SOC'], table['OCV(mV)'])
    return res # mV

# 温度敏感系数kt(soc)
def kt(soc, table1, table2, T1, T2):
    ocv1 = ocv_ref_23(soc, table1)
    ocv2 = ocv_ref_23(soc, table2)
    return (ocv2 - ocv1) / (T2 - T1) # mV/°C

# 综合OCV模型
def ocv(soc, t, soh):
    k_t = kt(soc, discharging_23c, discharging_45c, 23, 45)
    return ocv_ref_23(soc, discharging_23c) + k_t * (t - t_ref) 

def ocv(soc):
    return ocv_ref_23(soc, discharging_23c)

ocv(0.5)
```
